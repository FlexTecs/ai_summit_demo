We will be interacting with the terminal in a Google Colab-provisioned VM.


## Setting up Claude Code in the Colab terminal

1. Open up Terminal --> click on the Terminal button at the bottom (next to `{} Variables`)
2. Install Claude Code in terminal (copy after the `$`): ```curl -fsSL https://claude.ai/install.sh | bash```
3. Verify your installation with `claude --version`
  a. You should see a response of this format: `2.1.128 (Claude Code)`
4. Now, we are going to ensure that each group's Team A authenticates a Claude Pro subscription with this command in terminal: `claude`. You will see some variant of this output below:
```
Welcome to Claude Code v2.1.128
…………………………………………………………………………………………………………………………………………………………

     *                                       █████▓▓░
                                 *         ███▓░     ░░
            ░░░░░░                        ███▓░
    ░░░   ░░░░░░░░░░                      ███▓░
   ░░░░░░░░░░░░░░░░░░░    *                ██▓░░      ▓
                                             ░▓▓███▓▓░
 *                                 ░░░░
                                 ░░░░░░░░
                               ░░░░░░░░░░░░░░░░
       █████████                                        *
      ██▄█████▄██                        *
       █████████      *
…………………█ █   █ █………………………………………………………………………………………………………………
```
5. Select your theme (1. Auto is fine). Hit enter to continue.
6. Now Login with your credentials: For your method, select "1. Claude account with subscription"
7. It won't be able to open a browser. That's ok. copy the URL that starts wtih https://claude.com/cai/oauth... into your browser in a new tab.
8. In the new, click the Authorize button.
9. Copy the Authentication Code displayed in the browser
10. Navigate back to your Terminal tab and paste it in. Hit Enter.

It should say "Login Successful, press enter to continue"
